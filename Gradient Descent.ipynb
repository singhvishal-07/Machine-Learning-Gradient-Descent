# Gradient Descent in ML
## This helps us to find the equation for a given set of inputs and outputs
for a given equation : y= mx+b, we calculate the values of 'm' and 'b'

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=(20.0,10.0)
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 500
    n = len(x)
    learning_rate = 0.08

    # we will print the scatter points 
    plt.scatter(x,y,color = 'blue', marker = 'o')
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        # Here once we have calculated the y_predicted value 
        # We will now print the y_predicted lines
        plt.plot(x,y_predicted,color = 'red')
        
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n) * sum(x*(y-y_predicted))
        bd = -(2/n) * sum((y-y_predicted))
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        #print('m = %f, b = %f, cost = %f, iteration = %d'%(m_curr,b_curr,cost,i))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
print(x,y)
```

### The Graph below shows the y_predicted line and the scatter points 
It shows the best fit line with m = 2 and b =3 

```python
gradient_descent(x,y)
```

# Exercise for Gradient Descent :-
## Using GradientExercise.csv file
### finding the relation between 'Math' and 'CS'

```python
import pandas as pd
from sklearn import linear_model 
import numpy as np
data = pd.read_csv('GradientExercise.csv')
print(data.shape)
data.head(10)
```

```python
x = data['math'].values
y = data['cs'].values
x = x.astype(int)
y = y.astype(int)
x = x[:10]
y = y[:10]
print(x,y)
```

### FInding the intercepts using Linear Regression model

```python
# calculating the means of x and y
mean_x = np.mean(x)
mean_y = np.mean(y)

# Finding the total number of values in the dataset
n = len(x)

# calculating the b0 and b1 values using the formula

numer = 0
denom = 0

for i in range(n):
    numer += (x[i]-mean_x)*(y[i]-mean_y)
    denom += (x[i]-mean_x)**2
    
b1 = numer/denom
b0 = mean_y - (b1*mean_x)

# Printing the values of b0 and b1
print('m = %f, b = %f'%(b1,b0))
```

### Scatter Plot fro the given set of points and the best fit line :

```python
%matplotlib inline
plt.rcParams['figure.figsize']=(20.0,10.0)
# Plotting the regression line 

max_x = np.max(x) + 100
min_x = np.min(x) - 100

# Calculating line values x and y
a = np.linspace(min_x,max_x,1000)
b = b0 + b1*a

# Plotting the line
plt.plot(a,b,color='blue', label='Regression line')
# Plotting the scatter points
plt.scatter(x,y,c='red', label='Scatter points')

plt.legend()
plt.show()
```

### Finding the equation using the Gradient Descent

```python
%matplotlib inline
import math
#plt.rcParams['figure.figsize']=(20.0,10.0)
def gradient_desc(x,y):
    m_curr = b_curr = 0
    iterations = 100000
    n = len(x)
    learning_rate = 0.0002
    previous_cost = 0

    # we will print the scatter points 
    plt.scatter(x,y,color = 'blue', marker = 'o')
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        # We will now print the y_predicted lines
        plt.plot(x,y_predicted,color = 'red')
        
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n) * sum(x*(y-y_predicted))
        bd = -(2/n) * sum((y-y_predicted))
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        if math.isclose(cost, previous_cost , rel_tol=1e-20):
            break
        previous_cost = cost
        print('m = %f, b = %f, cost = %f, iteration = %d'%(m_curr,b_curr,cost,i))
```

```python
gradient_desc(x,y)
```
